```python
# meet 5
# BarChart
x = [2,4,6,8,10]
y = [6,7,8,2,4]

x2 = [1,3,5,7,9]
y2 = [23,42,32,25,15]

plt.bar(x,y,label = "bars1")
plt.bar(x2,y2,label = "bars2",color = "red")

plt.xlabel("x")
plt.ylabel("y")
plt.title("sample graph \nsimple plot")
plt.legend()
plt.show()
```

```python
# Histogram
population_ages = [22,55,62,42,4,99,21,22,55,63,45,26,36,95,25,20,14,15,20,12,32,78,62,74,84,51,65,20,24,48,57,95,15,35,20,48,94,93,74,75,71]

# ids = [x for x in range(len(population_ages))]
# plt.bar(ids,population_ages)

bins = range(0,110,10)
plt.hist(population_ages,bins,histtype = "bar",rwidth = 0.8)

plt.xlabel("x")
plt.ylabel("y")
plt.title("sample graph \nsimple plot")
plt.show()
```

```python
# meet 7

days = range(1,6)
sleeping = [7,8,6,9,7]
working = [7,8,5,9,6]
eating = [1,2,1,3,4]
playing = [3,2,1,2,3]

plt.plot([],[],color = "g",label = "sleeping" , linewidth = 5)
plt.plot([],[],color = "orange",label = "working", linewidth = 5)
plt.plot([],[],color = "r",label = "eating", linewidth = 5)
plt.plot([],[],color = "b",label = "playing", linewidth = 5)


plt.stackplot(days , sleeping , working , eating , playing , colors =["g","orange","r","b"])

plt.xlabel("x")
plt.ylabel("y")
plt.title("sample graph \nsimple plot")
plt.show()
```

```python
from datetime import date
from matplotlib import dates as mpl_dates

plt.style.use("seaborn")
dates = [
    date(2019,6,25),
    date(2019,8,23),
    date(2019,9,19),
    date(2020,3,10),
    date(2021,2,12),
    date(2021,9,16),
    date(2022,2,26)
]

y = [0,1,3,2,4,3,5]

plt.plot(dates,y)
plt.gcf().autofmt_xdate

date_format = mpl_dates.DateFormatter("%b,%d %y")
plt.gca().xaxis.set_major_formatter(date_format)
plt.show()
```

```python
# meet 10
import random
from itertools import count
from matplotlib.animation import FuncAnimation

plt.style.use("fivethirtyeight")

x_value = []
y_value = []
plt.plot(x_value,y_value)

index = count()
def animate(i):
    x_value.append(next(index))
    y_value.append(random.randint(0,5))
    plt.plot(x_value,y_value)

animated = FuncAnimation(plt.gcf() , animate,interval = 1000)
plt.tight_layout()
plt.show()
```
